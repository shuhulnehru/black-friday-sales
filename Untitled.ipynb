```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
train_df = pd.read_csv('train.csv')
train_df['source'] = 'train'
```

```python
test_df = pd.read_csv('test.csv')
test_df['source'] = 'test'
```

```python
df = pd.concat([train_df,test_df])
```

```python
df['City_Category'] = df['City_Category'].replace({'B':2,'C':1,'A':0})
```

```python
df['City_Category'].value_counts(normalize=True)
```

```python
df['Age'] = df['Age'].replace({"0-17":0 , "55+":1})
```

```python
def clean(num):
    if num==1:
        return 1
    
    elif num==0:
            return 0
    else:
        return 2
```

```python
df['Age'] = df['Age'].apply(clean)  #teenagers represent 0, senior people represent 1 and middle age group represent 2
```

```python
df['Age'].value_counts()
```

```python
df.info()
```

```python
df.drop(columns='Product_Category_3',inplace=True)
```

```python
df
```

```python
x = df['Product_Category_2'].median()
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(x)
```

```python
df.drop(columns=['User_ID','Product_ID'],inplace=True)
```

```python
df
```

```python
encoding = LabelEncoder()
```

```python
df['Gender'] = encoding.fit_transform(df.Gender)
```

```python
df['Stay_In_Current_City_Years'].value_counts()
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].replace({'4+':4})
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
```
